```python
import pickle
import os
import pandas as pd
```

```python
pickle_folder = '../models/'
pickles = os.listdir(pickle_folder)
```

```python
models = dict()
```

```python
for i in range(1,150):
    pickle_name = "station_{}_params.pkl".format(i)
    if pickle_name in pickles:
        model = pickle.load(open(pickle_folder+pickle_name, 'rb'))
        models[i] = model
```

```python
models_df = pd.DataFrame.from_dict(models, orient = 'index')
```
